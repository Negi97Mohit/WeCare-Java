```python
#20 Day moving average for the HIGH
ma_high_20=pd.DataFrame()
for i in range(row):
    if(((i+1)%20)==0):
        summer=pd.DataFrame(high.iloc[(i-19):i+1])
        ma=np.sum(summer)/20
        ma_high_20[i]=ma
    elif(i<19):
        ma_high_20[i]=0
    else:
     continue       
ma_high_20=ma_high_20.round(6).T


#20 Day moving average for the Open
ma_open_20=pd.DataFrame()
for i in range(row):
    if(((i+1)%20)==0):
        summer=pd.DataFrame(open.iloc[(i-19):i+1])
        ma=np.sum(summer)/20
        ma_open_20[i]=ma
    elif(i<19):
        ma_open_20[i]=0
    else:
     continue       
ma_open_20=ma_open_20.round(6).T


#20 Day moving average for the Low
ma_low_20=pd.DataFrame()
for i in range(row):
    if(((i+1)%20)==0):
        summer=pd.DataFrame(low.iloc[(i-19):i+1])
        ma=np.sum(summer)/20
        ma_low_20[i]=ma
    elif(i<19):
        ma_low_20[i]=0
    else:
     continue       
ma_low_20=ma_low_20.round(6).T

#20 Day moving average for the Close
ma_close_20=pd.DataFrame()
for i in range(row):
    if(((i+1)%20)==0):
        summer=pd.DataFrame(close.iloc[(i-19):i+1])
        ma=np.sum(summer)/20
        ma_close_20[i]=ma
    elif(i<19):
        ma_close_20[i]=0
    else:
     continue       
ma_close_20=ma_close_20.round(6).T

frame=[ma_open_20,ma_high_20,ma_low_20,ma_close_20]
ma_20=pd.concat(frame,ignore_index=True,sort=False)
ma_20
```

```python
#close train and test spliting model
#Using the LinearRegression
close1=gbpusd['Close']
print(close1.shape)
close1=pd.DataFrame(close1)
close1.shape
close_index=[]
row1,col1=gbpusd.shape
lm=LinearRegression()
close=pd.DataFrame(close)
for i in range(row1):
    low=i
    mid=low+100
    high=mid+100   
    if high<=(row1):
        X=close1[low:mid]
        y=close1[mid:high]
        X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.5,shuffle=False,random_state=0)
        lm.fit(X_train,y_train)
        y_predict=lm.predict(X_test)
        mse=mean_squared_error(y_test,y_predict)
        r2=r2_score(y_test,y_predict)
        if(r2>0.5):
            close_index.append(i)
            print("\nMSE",mse,"\n,R2:",r2)
            plt.scatter(X_test,y_test,color='black')
            plt.plot(X_test,y_predict,color='blue')
            plt.show()
print(close_index)
```
