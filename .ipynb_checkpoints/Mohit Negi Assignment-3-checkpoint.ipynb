```python
#Libraries Used
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn import metrics
from sklearn.datasets import load_boston
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_squared_error
```

### **1. Representing text as datam**

```python
#1.Representing text as data 

simple_train = ['call you tonight', 'Call me a cab', 'please call me... PLEASE!']
#simple_train=pd.DataFrame(simple_train,columns=['Master'])

#Learn the 'vocabulary' of the training data
vectorize=CountVectorizer()
vectorize.fit_transform(simple_train)
vectorize.get_feature_names()

#Transform training data into a 'document-term matrix' 
doc_term_matrix=vectorize.transform(simple_train)

#Print the sparse matrix.
print(doc_term_matrix)

#Convert the sparse matrix to a dense matrix
doc_term_matrix.toarray()

#Examine the vocabulary and document-term matrix together
pd.DataFrame(doc_term_matrix.toarray(),columns=vectorize.get_feature_names())

#Transform testing data into a document-term matrix
simple_test = ["please don't call me"]
doc_term_matrix=vectorize.transform(simple_test)
doc_term_matrix.toarray()

#Examine the vocabulary and document-term matrix together
pd.DataFrame(doc_term_matrix.toarray(),columns=vectorize.get_feature_names())
```

### **2.Analyzing SMS data with Naïve Bayes**

```python
#2.Analyzing SMS data with Naïve Bayes 
url='https://raw.githubusercontent.com/Negi97Mohit/Python_Na-ve-Bayes-/main/sms.tsv'
column_name= ['label', 'message']
master=pd.read_table(url,sep='\t',header=None,names=column_name)
print(master.head(10))

#Convert label to a numeric variable.
master.label.value_counts()
master['label']=master.label.map({'ham':0,'spam':1})

#Define X and y
X=master.message
y=master.label
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=1)
print("Shape of train set:",X_train.shape)
print("Shape of test set:",X_test.shape)

#Use the two Naïve Bayes models (GaussianNB and MultinomialNB) to finish the following requirements, respectively:
#Vectorizeing training dataset
X_train_doc_term_matrix=vectorize.fit_transform(X_train)
X_train_doc_term_matrix
#Test data to document term matrix
X_test_doc_term_matrix=vectorize.transform(X_test)
#Instanciating the Multinomia naive bayes model
nb=MultinomialNB()

#Training the model
nb.fit(X_train_doc_term_matrix,y_train)
#Making prediction
y_predict=nb.predict(X_test_doc_term_matrix)

#calculate accuracy of predictions
print("\nThe Accuracy of prediction is:",metrics.accuracy_score(y_test,y_predict))

#give the confusion matrix
print("\nThe confusion mtraix is:")
print(metrics.confusion_matrix(y_test,y_predict))

#print message text for the false positives
print("\nText for the false positives")
print(X_test[y_predict>y_test])

#print message text for the false negatives
print("\nText for the false negatives")
print(X_test[y_predict<y_test])
```

### **3.Linear Regression**

```python
#3. Linear Regression 

#First, read into the data and process it in the right format for using machine learning.
x=load_boston()
master=pd.DataFrame(x.data,columns=x.feature_names)
print(master.head())

#Then, use the DataFrame corr() method to show what features are correlated with each other.
plt.figure(figsize=(10,10))
correation=master.corr()
corr_lower=correation.where(np.tril(np.ones(correation.shape)).astype(np.bool))
print(corr_lower.info)
mask=np.triu(corr_lower)
sns.heatmap(corr_lower,annot=True,mask=mask,cmap='rainbow')


#The correlation coefficient has values between -1 to 1
# A value closer to 0 implies weaker correlation (exact 0 implying no correlation)
# A value closer to 1 implies stronger positive correlation
# A value closer to -1 implies stronger negative correlation
corr=corr_lower.abs()
high_corr=corr[corr>0.5]
print("\n Values with high correlation")
print(high_corr)
plt.show()

#Linear Regression
X_train,X_test,y_train,y_test=train_test_split(x.data,x.target)
linear=LinearRegression()
linear.fit(X_train,y_train)
prediction=linear.predict(X_test)
expected=y_test
coeff=pd.DataFrame(linear.coef_,index=['crim', 'zn', 'indus', 'chas', 'nox', 'rm', 'age', 'dis', 'rad', 'tax', 'ptratio', 'b', 'lstat'])
print("\n Prediction:")
print("\n",prediction,"\n")
print("\nmodel coefficient:",coeff)
print("\nmodel intercept:",linear.intercept_,"\n")
plt.figure(figsize=(4, 3))
plt.scatter(expected, prediction)
plt.plot([0, 50], [0, 50],'r')
plt.axis('tight')
plt.tight_layout()
plt.show()
print("\nThe R2 score for the model is :",r2_score(y_test,prediction),"\n")
print("\nThe mean squared error[MSE] for the model is :",mean_squared_error(y_test,prediction),"\n")
```
